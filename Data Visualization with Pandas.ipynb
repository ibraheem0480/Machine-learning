```python
df2.plot() #pandas by default as line grahp
```

```python
df2['bedrooms'].plot.line(figsize=(20,3))
```

```python
df2['bedrooms'].plot.line()
```

```python
df2['bedrooms'].plot.line(ls=':',c="black",lw=10) #ls mean line style
```

```python
title="bedroom graph"
xlabel="My x data"
ylabel="My y data"
```

```python
ax=df2['bedrooms'].plot.line(ls=':',c="red" ,title=title)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax=df2.plot()
ax.legend(loc=1)
```
